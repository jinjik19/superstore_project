```python
import pandas as pd
```

```python
def to_snake_case(name):
    name = name.strip()
    name = name.replace(' ', '_')
    return name.lower()
```

```python
orders_df = pd.read_excel('../data/Superstore.xls', sheet_name='Orders')
orders_df.rename(columns=to_snake_case, inplace=True)
orders_df.columns
```

```python
people_df = pd.read_excel('../data/Superstore.xls', sheet_name='People')
people_df.rename(columns=to_snake_case, inplace=True)
people_df.head(10)
```

```python
returns_df = pd.read_excel('../data/Superstore.xls', sheet_name='Returns')
returns_df.rename(columns=to_snake_case, inplace=True)
returns_df.head(10)
```
